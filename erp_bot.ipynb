````python
service = Service('path_to_chromedriver') 
options = Options() 
# options.add_argument("--headless")

options.add_argument('--profile-directory=Default') # Paste the Profile name from the profile path. Mine is "Default"
options.add_argument('--user-data-dir=C:\\Users\\Dnyandeep\\AppData\\Local\\Google\\Chrome\\User Data\\') #Paste Executable Path
````

````python

class ERPLogin:
    def __init__(self, login_id, password, father_name, mother_name, birthplace):
        self.login_id = login_id
        self.password = password
        self.browser = webdriver.Chrome(service=service, options=options)
        self.browser.maximize_window()
        self.link = "https://erp.iitkgp.ac.in/"
        self.browser.get(self.link)
        self.oldTab = self.browser.current_window_handle
        self.father_name = father_name
        self.mother_name = mother_name
        self.birthplace = birthplace

        time.sleep(1.5)
        
    def login(self):
        
        login_id_element = self.browser.find_element("xpath", '//*[@id="user_id"]')
        login_id_element.click()
        login_id_element.send_keys(self.login_id)
        password_element = self.browser.find_element("xpath", '//*[@id="password"]')
        password_element.click()
        password_element.send_keys(self.password)
        time.sleep(1.5)
        question_element = self.browser.find_element("xpath", '//*[@id="question"]').text
        if question_element == "Father's name":
            ans_element = self.browser.find_element("xpath", '//*[@id="answer"]')
            ans_element.click()
            ans_element.send_keys(self.father_name)
        elif question_element == "Mother's name":
            ans_element = self.browser.find_element("xpath", '//*[@id="answer"]')
            ans_element.click()
            ans_element.send_keys(self.mother_name)
        elif question_element == "Birthplace":
            ans_element = self.browser.find_element("xpath", '//*[@id="answer"]')
            ans_element.click()
            ans_element.send_keys(self.birthplace)
        
        self.browser.find_element("xpath", '//*[@id="getotp1"]').click()
        time.sleep(1.5)
        self.browser.switch_to.alert.accept()
    
    def get_otp(self):
        self.browser.switch_to.new_window('tab')
        link = "https://mail.google.com/"  #Change this line to set your default gmail account
        self.browser.get(link)
        OTP = ""

        def refresh_page():
            self.browser.refresh()
            time.sleep(1.8)

        refresh_page()
        refresh_page()
        refresh_page()
        refresh_page()
        refresh_page()
        refresh_page()


        page = self.browser.find_element("xpath", '//div[@class="Nr UI S2 vy"]') #description
        pageContent = str(page.text)
        pattern = "(The OTP for Sign In in ERP Portal of IIT Kharagpur is )(\d{6})"
        matches = re.findall(pattern, pageContent)
        OTP = str(matches[0][1].replace(" ", ""))

        print("OTP: ", OTP)
        self.browser.switch_to.window(self.oldTab)

        otp_box = self.browser.find_element("xpath", '//*[@id="email_otp1"]')
        otp_box.click() 
        otp_box.send_keys(OTP)

        sign_in = self.browser.find_element("xpath", '//*[@id="loginFormSubmitButton"]')
        sign_in.click()


#Enter your ERP details here! (one-time)
login_id = "<your user_name>"
password = "<your password>"
father_name = "<your fathers name>"
mother_name = "<your mothers name>"
birthplace = "<your birthplace>"

erp_login = ERPLogin(login_id, password, father_name, mother_name, birthplace)
erp_login.login()
erp_login.get_otp()
````
